# TRYING TO CLEAN UP

```python
truck_ds['Color'] =truck_ds['Color'].fillna('WHITE')
truck_ds['Engine_Manufacturer'] = truck_ds['Engine_Manufacturer'].fillna('CUMM')
truck_ds['Channel'] = truck_ds['Channel'].fillna('Retail')
truck_ds['Engine_Type'] = truck_ds['Engine_Type'].fillna('IN LINE') 
```

```python
for a_mean in numerical:
    print(a_mean, truck_ds[a_mean].dropna().mean())
    truck_ds[a_mean] = truck_ds[a_mean].fillna(round(truck_ds[a_mean].dropna().mean(),0))
```

```python
## Should we eliminate the List price or the Book value or the Pct_of_List value
#### DO IT!!!!!
# Remove list price and percent of list only. Keep book value
truck_ds = truck_ds.drop(columns=['Pct_of_List', 'List_Price'], axis = 1)
```

```python
# Delivery location has too large of a distribution, drop age_yearss beacues stuff does nnot aliign mathmatically duet to time as well
#### Half of the transmission types were Auto and half manual. therefore drop data???
truck_ds = truck_ds.drop(columns= ['Model', 'Front_Tire_Size', 'Rear_Tire_Size','Color', 'Delivery_Location', 'Age_Yrs', 'Transmission_Type', 'Model','Truck_Manufacturer'], axis=1)
```

```python
def _str(shit):
    return str(shit)
    
truck_ds['Year'] = truck_ds['Year'].apply(_str)
```

# Lets do some one hot encoding

```python
numerical = ['Book_Value', 'Model_Year', 'Odometer', 'Wheelbase_Size', 'Gear_Ratio', 'Num_Axles', 'Num_Wheels', 'Rear_Axle_Weight', 'front_axle_wt', 'Horsepower', 'Engine_RPM', 'eng_est_life']
```

```python
price_scaler = StandardScaler()
price_scaler.fit(truck_ds[numerical])
truck_ds[numerical] = price_scaler.transform(truck_ds[numerical])
```

```python
categorical = [var for var in truck_ds.columns if truck_ds[var].dtype=='O']
```

```python
encoder = OneHotEncoder(sparse =  False)
encoder.fit(truck_ds[categorical])
categorical_transformed = encoder.transform(truck_ds[categorical])
```

```python
truck_ds = truck_ds.drop(labels=categorical, axis=1)
out = np.concatenate(encoder.categories_).ravel().tolist()
one_hot_encoded_categorical_pd = pd.DataFrame(categorical_transformed, columns = out)
```

```python
fully_cleaned_data = pd.concat([truck_ds,one_hot_encoded_categorical_pd], axis= 1)
```
